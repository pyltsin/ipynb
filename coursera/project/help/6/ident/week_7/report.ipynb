# <center> Capstone проект №1. Идентификация пользователей по посещенным веб-страницам

## <center> 1. Цели и задачи проекта, описание исходных и обработанных данных и признаков

В этом проекте мы будем решать задачу идентификации пользователя по его поведению в сети Интернет.

Дана последовательность из нескольких веб-сайтов, посещенных подряд один и тем же человеком, мы будем идентифицировать этого человека. Идея такая: пользователи Интернета по-разному переходят по ссылкам, и это может помогать их идентифицировать (кто-то сначала в почту, потом про футбол почитать, затем новости, контакт, потом наконец – работать, кто-то – сразу работать).

Для каждого пользователя заведен csv-файл с названием user\*\*\*\*.csv (где вместо звездочек – 4 цифры, соответствующие ID пользователя), а в нем посещения сайтов записаны в следующем формате: <br>

<center>*timestamp, посещенный веб-сайт*</center>

Так как работать с именами сайтов не очень удобно, сформируем словарь сайтов, в котором ключом будет сайт, а значением его уникальный номер <span style="color:green">*uid*</span>, причем чем чаще сайт встречается, тем меньше его <span style="color:green">*uid*</span>.<br>

**Поставим задачу классификации: идентифицировать пользователя по сессии из <span style="color:red">*n*</span> подряд посещенных сайтов. Объектом в этой задаче будет сессия из <span style="color:red">*n*</span> сайтов, последовательно посещенных одним и тем же пользователем, признаками – индексы этих <span style="color:red">*n*</span> сайтов . Целевым классом будет id пользователя.**

Для классификации создадим матрицу признаков следующим образом:<br>
строкам будут соответствовать сессии из <span style="color:red">*n*</span> сайтов, а столбцам – индексы сайтов. На пересечении строки $i$ и столбца $j$ будет стоять число $n_{ij}$ – cколько раз сайт $j$ встретился в сессии номер $i$

Для анализа и построения модели мы будем использовать выборку из 10 пользователей, а тренировать и проверять модель на выборке из 150 пользователей.

## <center> 2. Первичный анализ признаков

### Анализ и модификация признаков 

Для 10 пользователей количество уникальных сайтов составило **4913** сайтов, а для 150 - **27797**.

Будем рассматривать перекрывающиеся сессии, добавим еще один параметр ширину окна <span style="color:red">*k*</span>. 

**Пример**: для длины сессии 10 и ширины окна 7 файл из 30 записей породит 5 сессий (1-10, 8-17, 15-24, 22-30, 29-30). При этом в предпоследней сессии будет один ноль, а в последней – 8 нолей.

Построим гистограмму распределения количества уникальных сайтов за сессию для 10 позьзователей
![alt](hist_uniq.png)
По ее виду можно предположить, что распределение не является нормальным.

Проверим наше предположение, построив Q-Q график и применим к выборке критерий Шапиро-Уилкса.
![alt](QQ.png)
p - value критерия равно 0, поэтому смело отвергаем гипотезу о нормальном распределении уникального количества сайтов за сессию.



Продолжим дальше исследовать наш признак - количество уникальных сайтов за сессию.
Проверим гипотезу о том, что доля сессий с уникальными сайтами равна 95%, то есть

$H_0$: доля количество сайтов засессию меньше 10 равна 0.95

$H_1$: количество сайтов за сессию равно 10 больше 0.95

Гипотеза $H_0$ отвергается на уровне значимости $\alpha=0.05$, отсюда можно сделать вывод, что на уровне значимости $0.05$ доля сессий с уникальными сайтами больше 95%.

Оценим частоту поялвения сайта в выборке: $95\%$ доверительный интервал, построенный по bootstrap-выборке равен 
$(22.515 35.763)$.

### Вывод:
* Для малого количества пользователей(несколько десятков) линейные методы позволяют достаточно хорошо определить пользователя по набору его сессий.

* Для большого количества пользователей(больше сотни) мы можем определять для каждого конкретного пользователя по его сессиям он это или нет, то есть возможноть определить злоумышленника, который действует от лица пользователя.

* Возможно следует для большего числа пользователей провести кластеризацию и определять не конкретного пользователя, а его принадлежность к соответсвующей группе.

* В данной задаче хорошо работаю линейные методы (в том числе и стохастические), возможно применение WV или Spark. Градиентный бутинг в данной задаче не показал сильного превоходства перед линейными методами.
