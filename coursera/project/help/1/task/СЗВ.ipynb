```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Показательное распределение
https://en.wikipedia.org/wiki/explace_distribution

Имеет плотность распределения $f(x)=\lambda e ^{-\lambda x}$

матожидание $\lambda ^{-1}$

дисперсию $\lambda ^{-2}$

lamb - $\lambda^{-1}$

```python
lamb = 2 #
```

```python
exp = sts.expon(scale = lamb)
```

Вычисляем среднее и дисперсию

```python
mean, var = exp.stats(moments='mv')
print(mean, var)
```

Генерируем выборку длины 1000

```python
sample = exp.rvs(size=1000)
```

Гистограмма выборки и график теоретической плотности распределения:

```python
x = np.linspace(0,6,1000)
plt.hist(sample, bins=30 , normed=True, label='hist', color="green")
pdf = exp.pdf(x)
plt.plot(x, pdf, label='theoretical pdf',color="red")
plt.legend()

plt.ylabel('f(x)')
plt.xlabel('$x$')
```

## Гистограмы выборочного среднего.

1000 выборок длины 3.
Нормальная модель имеет параметры $(\lambda^{-1},\lambda^{-1}\sqrt{\frac{1}{3}})$

```python
X_3 = [exp.rvs(3).mean() for i in range(1000)]
norm_rv = sts.norm(loc=mean, scale=np.sqrt(var/5))
x = np.linspace(0,6,100)
pdf = norm_rv.pdf(x)

plt.hist(X_3,normed=True, bins=30, label='hist',color="green")
plt.plot(x, pdf, label='theoretical pdf',color="red")
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend()
```

1000 выборок длины 5.
Нормальная модель имеет параметры $(\lambda^{-1},\lambda^{-1}\sqrt{\frac{1}{5}})$

```python
X_5 = [exp.rvs(5).mean() for i in range(1000)]
norm_rv = sts.norm(loc=mean, scale=np.sqrt(var/5))
x = np.linspace(0,6,100)
pdf = norm_rv.pdf(x)

plt.hist(X_5,normed=True, bins=30, label='hist',color="green")
plt.plot(x, pdf, label='theoretical pdf',color="red")
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend()
```

1000 выборок длины 10.
Нормальная модель имеет параметры $(\lambda^{-1},\lambda^{-1}\sqrt{\frac{1}{10}})$

```python
X_10 = [exp.rvs(10).mean() for i in range(1000)]
norm_rv = sts.norm(loc=mean, scale=np.sqrt(var/10))
x = np.linspace(0,5,100)
pdf = norm_rv.pdf(x)

plt.hist(X_10,normed=True, bins=30, label='hist',color="green")
plt.plot(x, pdf, label='theoretical pdf',color="red")
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend()
```

1000 выборок длины 50.
Нормальная модель имеет параметры $(\lambda^{-1},\lambda^{-1}\sqrt{\frac{1}{50}})$

```python
X_50 = [exp.rvs(50).mean() for i in range(1000)]
norm_rv = sts.norm(loc=mean, scale=np.sqrt(var/50))
x = np.linspace(1,3.2,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.hist(X_50,normed=True, bins=25, label='hist')
plt.plot(x, pdf, label='theoretical pdf')
plt.ylabel('f(x)')
plt.xlabel('$x$')
plt.legend()
```

### Вывод:
Для длин выборок 3 и 5 - гистограммы выброк не совпадают с графиками плотности нормального распределения. При длине выборки больше или равной 10, мы видим лучшее приближение плотности нормального распределения гистограммой.
