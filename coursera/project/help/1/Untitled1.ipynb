```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import exp, sqrt
%matplotlib inline
```

# Логнормальное распределение
Сначала строим выборку по логнормальному распределению и сравниваем гистограмму с теоретической плотностью для логнормального распределения

```python
mu_lognormal =0.3
sko_lognormal = 1.4
lognorm_rv = sts.lognorm(mu_lognormal, sko_lognormal)
sample = lognorm_rv.rvs(1000)
plt.hist(sample, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
x = np.linspace(1,5,100)
pdf = lognorm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
```

Вполне похожи. А теперь строим гистограммы распределения выборочных средних. Начнем с n=5. Для каждой выборки будем записывать среднее в массив. Но для начала посчитаем какое нормальное распределение будет соответствовать гистограммам.

```python
mu_normal = lognorm_rv.mean()
disp_lognormal = lognorm_rv.var()
print mu_normal, disp_lognormal
```

```python
sko_normal = sqrt(disp_lognormal/5.0)
mean_list=[]
i=0
while i<1000:
    lognorm_rv = sts.lognorm(mu_lognormal, sko_lognormal)
    sample = lognorm_rv.rvs(5)
    mean_list.append(sample.mean())
    i=i+1
plt.hist(mean_list, normed=True)
plt.ylabel('number of means')
plt.xlabel('$x$')
x = np.linspace(1,4,100)
norm_rv = sts.norm(mu_normal, sko_normal)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
```
