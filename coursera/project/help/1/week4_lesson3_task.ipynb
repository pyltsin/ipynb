```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import sqrt
%matplotlib inline
import random
```

# Приближение выборочных средних Показательного (экспоненциального) распределения по ЦПТ

$f(x)$ - функция плотности показательного распределения случайной величины $X$ ~ $EXP(\lambda)$

$f(x) = 0, если \quad x < 0 $

$f(x) = \lambda e^{-\lambda x}, если \quad x \ge 0 , где \quad параметр \quad \lambda > 0$

В нашем примере возьмем $\lambda = 0.5$, для этого на вход функций Питона будем подавать $\beta = 2.0$, являющийся обратным к $\lambda$

Матожидание $EX = 1/\lambda = 2$

Дисперсия $DX = 1/\lambda^2 = 4$

Генерируем выборку из 1000 элементов случайной величины X ~ EXP(0.5) и строим ее гистограмму. 
На этом же графике строим теоретическую функцию плотности вероятности показтельного распределения



```python
beta = 2.0
sample = np.random.exponential(beta,1000)
plt.hist(sample,bins = 30, label = 'histogram X ~ EXP(0.5)', normed = True)
plt.ylabel('Number of samples')
plt.xlabel('$x$')

x = np.linspace(-1,10,100)
rv = sts.expon.pdf(x, loc = 0., scale = beta)
plt.plot(x, rv, label='theoretical PDF X ~ EXP(0.5)')
plt.legend(loc='upper right')
plt.show()
```

сгенерируем выборку из 1000 элементов объема n = 5



```python
n = 5
m = 1000
#генерируем массив n x m, каждая строка - выборка из m элементов
sample = np.array([np.random.exponential(beta,m) for index in range(n)])
#зададим массив sample_mean размером 1 x m как среднее от элементов sample
sample_mean = np.array([0. for i in range(m)]) 
#сначала зададим массив как нулевой, затем возьмем сумму от элементов массива sample поэлементно
for i in range(0,n):
    sample_mean += sample[i]

sample_mean = sample_mean/n #усредняем по n

plt.hist(sample_mean,bins = 20, label = 'histogram Xn ~ EXP(0.5)', normed = True)
plt.ylabel('Number of samples')
plt.xlabel('$x$')

x = np.linspace(-1,8,m)
norm_rv = sts.norm(beta, sqrt((beta**2)/n))
pdf = norm_rv.pdf(x)
plt.plot(x,pdf,label='normal PDF N(2,sqrt(4/n)')
plt.legend(loc='upper right')
plt.show()
```
