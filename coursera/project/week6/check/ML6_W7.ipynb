# Финальный проект специализации: резюме

## 1. Цели и задачи проекта, описание исходных и обработанных данных и признаков

Основной задачей проекта является идентификация пользователя по его поведению в сети Интернет.

Имеются данные с прокси-серверов Университета Блеза Паскаля, они имеют очень простой вид. Для каждого пользователя заведен csv-файл с названием user\*\*\*\*.csv (где вместо звездочек – 4 цифры, соответствующие ID пользователя), а в нем посещения сайтов записаны в следующем формате: <br>

<center>*timestamp, посещенный веб-сайт*</center>

Для дальнейшей работы данные переформатируются следующим образом: в каждой строке мы имеет сессию пользователя X с временем посещения сайта и наименованием сайта. Для обучения далее мы будем использовать разреженную матрицу таблицы этих данных.

## 2. Первичный анализ признаков

- С помощью теста Шапиро-Уилка было выяснено, что распределение числа уникальных сайтов в сессии нельзя назвать нормальным.
- С помощью биномиального теста проверена гипотеза о том, что доля случаев, когда пользователь повторно посетил какой-то сайт (то есть число уникальных сайтов в сессии < 10) велика, и найден соответствующий доверительный интервал.
- Построен доверительный интервал для средней частоты появления сайта в выборке.

## 3. Первичный визуальный анализ признаков

В ходе первичного визуального анализа признаков были построены следующие гистограммы:

- Гистограммы длины сессии
- Гистограммы числа уникальных сайтов в сессии
- Гистограммы часа начала сессии 
- Гистограммы дня недели начала сессии 
- и прочие.

Для разных пользователей гистограммы заметно отличаются, что даёт возможность построить новые релевантные признаки.

## 4. Дополнительные признаки

Дополнительные признаки, которые были добавлены в данные:
    
- Всевозможные временные признаки: час/день/неделя/месяц начала сессии, час конца сессии.
- Длительность сессии
- Наличие в сессии популярных сайтов (facebook/google/youtube)
- Среднее время пребывания на сайте

## 5. Оценка модели

Для построения модели использовалась стратифицированная кросс-валидации и следующие алгоритмы (классификаторы):
- K-Neighbors
- Random Forest
- Logistic Regression
- LightGBM/XGBoost

Параметры для каждого из алгоритмов (константа регуляризации, глубина/количество деревьев для Random Forest/бустинга) подбирались по сетке с помощью GridSearch.

Метрики, с помощью которых оценивалось качество алгоритмов: accuracy и ROC AUC (по большей части).
После подбора параметров лучшие результаты показали Logistic Regression и LightGBM/XGBoost (с небольшой разницей).
На тестовой выборке метрика ROC AUC достигает 0.948 (для Random Forest и бустинга значение метрики несущественно ниже).

## 6. Вывод

Построена модель, которая даёт возможность решать задачу идентификации пользователя в сети Интернет с достаточно высокой точностью.
