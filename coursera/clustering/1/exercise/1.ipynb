```python
df_out['label'] = cluster.labels_.tolist()
```

```python
gb = df_out.groupby('label')
```

```python
inm = gb.count()>15
```

```python

inm['label'] = inm.index
```

```python
inm = inm[['latitude', 'label']]
```

```python
inm.columns = ['tr', 'label']
```

```python
inm.head(1)
```

```python
cluster.cluster_centers_.shape
```

```python
cluster.cluster_centers_
```

```python
inm['latitude'] = pd.Series(cluster.cluster_centers_[:,0])
inm['longitude'] = pd.Series(cluster.cluster_centers_[:,1])
```

```python
list_center = [[33.751277, -118.188740],[25.867736, -80.324116],[51.503016, -0.075479],
[52.378894, 4.885084],
[39.366487, 117.036146],
[-33.868457, 151.205134]]
```

```python
def dim_row(row):
    l = []

    x_x = row['latitude']
    y_y = row['longitude']
    for x, y in list_center:
        r = ((x_x-x)**2+(y-y_y)**2)
        l.append(r)
    return min(l)
```

```python
inm['dim'] = inm.apply(dim_row, axis = 1)
```

```python
inm[inm['tr']].sort_values(by='dim')[0:20]
```
