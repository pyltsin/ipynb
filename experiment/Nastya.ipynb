Можно сделать так:

```python
def parabolla(X):
    y =[]
    for x in X:
        y.append(x**2+2*x+1)
    return y
```

Если работать с массивами можно сделать так (почти похоже, но скорость на больших данных в сотни раз больше)

```python
def par_np(X):
    return np.power(X, 2) +2 * X + 1
```

```python
x = np.arange(-20,20, 0.1)
```

```python
y = par_np(x)
```
